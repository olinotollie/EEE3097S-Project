```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy import fftpack
data= pd.read_csv("TT_data.csv")
data
```

```python
t = data.Time
A_x=data.Accel_X
A_y=data.Accel_Y
A_z=data.Accel_Z
```

```python
plt.plot(t, A_x)
plt.title("Accelerometer data in X-axis")
plt.xlabel("time")
plt.ylabel("accelerometer X (g)")
```

```python
fs=2048
N=47667
fstep= fs/N
A_X_fft=np.fft.fft(A_x)
A_X_mag= np.abs(A_X_fft)/N
A_X_mag_plot= 2*A_X_mag[0:int(N/2+1)]
A_X_mag_plot[0]= A_X_mag[0]/2
freq=np.linspace(0, (N-1)*fstep, N)
f_plot= freq[0:int(N/2+1)]
plt.plot(f_plot, A_X_mag_plot)
plt.title("FFT of accelerometer data in X-axis")
plt.xlabel("frequency")
plt.ylabel("magitude of readings")
```

```python
plt.plot(t, A_y)
plt.title("Accelerometer data on the Y-axis")
plt.xlabel("time")
plt.ylabel("accelerometer Y (g)")
```

```python
A_Y_fft=np.fft.fft(A_y)
A_Y_mag= np.abs(A_Y_fft)/N
plt.plot(freq, A_Y_mag)
plt.title("FFT of accelerometer data on the Y-axis")
plt.xlabel("frequency")
plt.ylabel("magitude of readings")
```

```python
plt.plot(t, A_z)
plt.title("Accelerometer data on the Z-axis")
plt.xlabel("time")
plt.ylabel("accelerometer Z (g)")
```

```python
A_Z_fft=np.fft.fft(A_z)
A_Z_mag= np.abs(A_Z_fft)/N
plt.plot(freq, A_Z_mag)
plt.title("FFT of accelerometer data on the Z-axis")
plt.xlabel("frequency")
plt.ylabel("magitude of readings")
```
